```python
dfii.describe()
```

```python
dfii.describe()
```

The '.describe' command did return more information, because  '2018 Mean Annual Wage' variable is a 'object', central tendancy measures are not returned. Below the '2018 Mean Annual Wage' is chnaged to a 'float'

```python
dfii["2018 Mean Annual Wage"].isna()
dfii["2018 Mean Annual Wage"].info()
dfii["2018 Mean Annual Wage"] = dfii["2018 Mean Annual Wage"].str.replace('[\$,]', '', regex=True).astype(float)
dfii["2018 Mean Annual Wage"].info()
```

```python
dfii.describe()
```

Convcerting to 'float' worked.  The final dataset has 51 rows and 2 columns. Mean annual wage = $87,563.33 with a minimum value of $75,010 and a max value of $107,920.

The describe command for this particular dataset is not particularlly useful because most of the data is address data and therefore not useful for central tendencies.  The exceptions maybe for '# of Dwelling Units' and 'Infested Dwelling Unit Count'.  '# of Dwelling Units' may be useful because it provides potential information about poulation density, while 'Infested Dwelling Unit Count' provides information about the number of units infested in a building.

```python
#below i perform some minor wrangling to provide more information about the data
Buildings_infestation_counts = dfiv.groupby('Building ID')['Infested Dwelling Unit Count'].count()
Buildings_infestation_counts = Buildings_infestation_counts.sort_values(ascending=False)
df_counts = pd.DataFrame(Buildings_infestation_counts)
df_counts.reset_index(inplace=True)
df_counts.columns = ['Building ID', 'Infested Dwelling Unit Count']
print(df_counts)
```
